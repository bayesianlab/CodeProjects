```python
import collections 

class Solution(object):

    def __init__(self) -> None:
        pass 
    

    def fulldict(self, letters):
        if('a' in letters.keys() and 'b' in letters.keys() and 'c' in letters.keys()):
            return True
        else:
            return False 
        
    def minmax(self, letters):
        maxa = max(letters['a'])
        maxb = max(letters['b'])
        maxc = max(letters['c'])
        return min(maxa,maxb,maxc)
    
    def maxmin(self, letters):
        maxa = min(letters['a'])
        maxb = min(letters['b'])
        maxc = min(letters['c'])
        return max(maxa,maxb,maxc)
                
    def numberOfSubstrings(self, s):
        """
        :type s: str
        :rtype: int
        """
        letters = {}
        i = 0
        swept = 0 
        nsubs = 0
        while(i < len(s)):
            if(s[i] not in letters.keys()):
                letters[s[i]] = [i]
            else:
                letters[s[i]].append(i)
            if(self.fulldict(letters)):                
                mm = self.minmax(letters)
                mx = self.maxmin(letters)
                biggerthancur = len(s) - 1 - mx 
                added = biggerthancur
                if(swept == 0):
                    swept += 1
                    added += mm 
                    multiplied = mm * biggerthancur
                    nsubs += 1 + multiplied + added 
                    letters.pop(s[mm])
                    lastmm = mm 
                else:
                    if(mm-lastmm == 1):
                        lastmm = mm
                        nsubs += 1 + added 
                    else:
                        added += (mm-lastmm-1)
                        multiplied = (mm-lastmm-1)*biggerthancur
                        lastmm = mm
                        nsubs += 1 + added + multiplied
                    letters.pop(s[mm])      
            i += 1
        return nsubs

    def numberOfSubstrings2(self, s: str) -> int:
        j, result, counter = 0, 0, collections.Counter()
        for i, c in enumerate(s):
            counter[c] += 1
            while len(counter) == 3:
                result += len(s) - i
                counter[s[j]] -= 1
                if not counter[s[j]]: del counter[s[j]]
                j += 1
        return result
        
        
sol = Solution()
print((sol.numberOfSubstrings('abab'), 0))
print((sol.numberOfSubstrings('abc'), 1))
print((sol.numberOfSubstrings('aabcc'), 4))
print((sol.numberOfSubstrings('aabcac'), 8))
print((sol.numberOfSubstrings('aabcc'), 4))
print((sol.numberOfSubstrings('aaabc'), 3))
print((sol.numberOfSubstrings('abcaabbabc'), 25))
print((sol.numberOfSubstrings('abcabc'),10))
print((sol.numberOfSubstrings('abbbbac'), 5))
print((sol.numberOfSubstrings('ababbcabc'), 21))
print((sol.numberOfSubstrings("ababbbc"),3))
print((sol.numberOfSubstrings("acbcac"), 8))
import time
s = time.time()
print((sol.numberOfSubstrings2("aaaaaaaaaaaaaabbbbbbbbbbcccccccccccabcabcbaaabbcbcbabbbcbabbabbcb"), 1549))
e = time.time()
print(e-s)
s = time.time()
print((sol.numberOfSubstrings("aaaaaaaaaaaaaabbbbbbbbbbcccccccccccabcabcbaaabbcbcbabbbcbabbabbcb"), 1549))
e = time.time()
print(e-s)
```
