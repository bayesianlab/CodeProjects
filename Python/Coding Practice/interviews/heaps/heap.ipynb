```python
class MaxHeap():
    def __init__(self) -> None:
        self.heap = []
        self.heap_sz = 0
    
    def push(self, val):
        
        if(len(self.heap)==0):
            self.heap.append(val)
            self.heap_sz = 1
            return 
        self.heap.append(val)
        self.heap_sz = len(self.heap)
        self.moveup(len(self.heap) - 1)
    
    def moveup(self, end):
        near = (end-1)//2
        if(near >=0 and (self.heap[near] < self.heap[end])):
            t = self.heap[near] 
            self.heap[near] = self.heap[end]
            self.heap[end] = t
            self.moveup(near)
        else:
            return 
        
    def heapify(self, arr):
        idx = 0
        lft = 1
        rgt = 2
        self.heap = arr 
        self.heap_sz = len(arr)
        while(lft < len(arr) or rgt < len(arr)):
            if(lft< len(arr) and rgt < len(arr)):
                if(arr[lft] > arr[rgt]):
                    self.moveup(lft) 
                else:
                    self.moveup(rgt)
            else:
                self.moveup(lft)
            idx += 1
            lft = 2*idx + 1
            rgt = 2*idx + 2
            
    def popmax(self):
        maxelem = self.heap[0]
        if(self.heap_sz == 1):
            return self.heap.pop()
        else:
            self.heap[0] = self.heap[self.heap_sz -1]
            del self.heap[self.heap_sz-1]
            self.heap_sz = len(self.heap)
            self.heapify(self.heap)
        return maxelem
        

def isMaxHeap(max_heap):
    idx = 0
    lft = 1
    rgt = 2
    max_heap_sz = len(max_heap)
    while(lft < len(max_heap) or rgt < len(max_heap)):
        if(lft< len(max_heap) and rgt < len(max_heap)):
            if(max_heap[idx] < max_heap[lft] and max_heap[idx] < max_heap[rgt]):
                return False 
        idx += 1
        lft = 2*idx + 1
        rgt = 2*idx + 2 
    return True      
        
    

            
        
    
# H = MaxHeap()
# H.push(1)
# H.push(2)
# H.push(4)
# H.push(7)
# print(H.heap)
H1 = MaxHeap()
H1.heapify([1,2,4,6])
H1.push(10)
print(H1.heap)
print(H1.popmax())
print(H1.popmax())
print(H1.heap)
print(isMaxHeap(H1.heap))
notheap = [1,4,4,21,4,62,0]
print(isMaxHeap(notheap))
```
