```python
def findroot(x, g, ep):
    a = 0
    b = g
    c = 1
    while(abs(g*g - x) > ep ):
        if(g*g >x):
            b = g
            g = (b+a)/2.
        else:
            a = g
            g = (b+a)/2.
        # print((c,abs(g*g-x)))
        c+=1
    return g

def findroot_iterations(x, g, ep):
    a = 0
    b = g
    c = 1
    while(abs(g*g - x) > ep ):
        if(g*g >x):
            b = g
            g = (b+a)/2.
        else:
            a = g
            g = (b+a)/2.
        # print((c,abs(g*g-x)))
        c+=1
    return c 


import numpy as np 
import matplotlib.pyplot as plt 

X = np.linspace(1,1000)
roots_iterations = [findroot_iterations(x,x,.01) for x in X ]
logn = [(1/np.log10(2))*np.log10(100*x)  for x in X]

plt.plot(X, logn)
plt.plot(X, roots_iterations)
```
