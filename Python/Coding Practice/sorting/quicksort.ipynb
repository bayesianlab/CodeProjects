```python
import numpy as np


def pivot(arr):
    r = len(arr)-1
    l = 0
    p = (r+l)//2
    piv = arr[p]
    while(l < r):
        if(piv > arr[r]):
            t = arr[r]
            arr[r] = arr[l]
            arr[l] = t
            if(l == p):
                p = r
            l += 1
        elif(piv < arr[l]):
            t = arr[l]
            arr[l] = arr[r]
            arr[r] = t
            if(r == p):
                p = l
            r -= 1
        else:
            l += 1
            r -= 1
            if(l >p):
                l=p
            if(r<p):
                r=p
    return p


def quicksort(arr, left, right):
    if(len(arr) == 2):
        if(arr[0] > arr[1]):
            arr = [arr[1], arr[0]]
            return arr
        else:
            return arr
    elif(len(arr) < 2):
        return arr
    else:
        part = pivot(arr)
        arr[0:part] = quicksort(arr[0:part], 0, part)
        arr[(part+1):right+1] = quicksort(arr[(part+1):right+1],
                                        part+1, len(arr[(part+1):right+1])-1)
        return arr



np.random.seed(1)
x = list((np.random.uniform(0, 1, 8)*100).round())
y = x[:] 
if(sum([i-j for i,j in zip(quicksort(x,0,len(x)-1), sorted(y))]) != 0):
    print("FAILED")
else:
    print("PASSED")
x = list((np.random.uniform(0, 1, 100)*100).round())
y = x[:] 
if(sum([i-j for i,j in zip(quicksort(x,0,len(x)-1), sorted(y))]) != 0):
    print("FAILED")
else:
    print("PASSED")
x = list((np.random.uniform(0, 1, 1000)*1000).round())
y = x[:] 
if(sum([i-j for i,j in zip(quicksort(x,0,len(x)-1), sorted(y))]) != 0):
    print("FAILED")
else:
    print("PASSED")
x = list((np.random.uniform(0, 1, 10000)*1000).round())
y = x[:] 
if(sum([i-j for i,j in zip(quicksort(x,0,len(x)-1), sorted(y))]) != 0):
    print("FAILED")
else:
    print("PASSED")
x = list((np.random.uniform(0, 1, 3)*1000).round())
y = x[:] 
if(sum([i-j for i,j in zip(quicksort(x,0,len(x)-1), sorted(y))]) != 0):
    print("FAILED")
else:
    print("PASSED")
x = list((np.random.uniform(0, 1, 2)*1000).round())
y = x[:] 
if(sum([i-j for i,j in zip(quicksort(x,0,len(x)-1), sorted(y))]) != 0):
    print("FAILED")
else:
    print("PASSED")
x = list((np.random.uniform(0, 1, 1)*1000).round())
y = x[:] 
if(sum([i-j for i,j in zip(quicksort(x,0,len(x)-1), sorted(y))]) != 0):
    print("FAILED")
else:
    print("PASSED")
x = list((np.random.uniform(0, 1, 0)*1000).round())
y = x[:] 
if(sum([i-j for i,j in zip(quicksort(x,0,len(x)-1), sorted(y))]) != 0):
    print("FAILED")
else:
    print("PASSED")
x = list((np.random.uniform(0, 1, 5)*1000).round())
y = x[:] 
if(sum([i-j for i,j in zip(quicksort(x,0,len(x)-1), sorted(y))]) != 0):
    print("FAILED")
else:
    print("PASSED")
x = list((np.random.uniform(0, 1, 20)*1000).round())
y = x[:] 
if(sum([i-j for i,j in zip(quicksort(x,0,len(x)-1), sorted(y))]) != 0):
    print("FAILED")
else:
    print("PASSED")
x = list((np.random.uniform(0, 1, 500)*1000).round())
y = x[:] 
if(sum([i-j for i,j in zip(quicksort(x,0,len(x)-1), sorted(y))]) != 0):
    print("FAILED")
else:
    print("PASSED")


```
