```python
from collections import deque


class Node:
    def __init__(self, v) -> None:
        self.data = v
        self.left = None
        self.right = None


class Tree:
    def __init__(self) -> None:
        self.root = None

    def addLeaf(self, d):
        if(self.root == None):
            self.root = Node(d)
            return
        r = self.root
        c = r
        while(1):
            if(d < r.data):
                if(r.left == None):
                    r.left = Node(d)
                    return
                else:
                    r = r.left
            else:
                if(r.right == None):
                    r.right = Node(d)
                    return
                else:
                    r = r.right

    def printTree(self):
        print(self.root.data)
        root = self.root
        self.root = root.left
        if(self.root is not None):
            self.printTree()
        self.root = root.right
        if(self.root is not None):
            self.printTree()
        return

    def calcHeight(self):
        nodes = [self.root]
        n2 = []
        h = 0
        while(len(nodes) > 0):
            for i in nodes:
                if(i.left):
                    n2.append(i.left)
                if(i.right):
                    n2.append(i.right)
            h+=1
            nodes = n2.copy()
            n2.clear()
        return h 

def maxDepth(node):
    if node is None:
        return 0 ; 

    else :

        # Compute the depth of each subtree
        lDepth = maxDepth(node.left)
        rDepth = maxDepth(node.right)
        print(lDepth)
        print(rDepth)
        print()

        # Use the larger one
        if (lDepth > rDepth):
            return lDepth+1
        else:
            return rDepth+1



t = Tree()
t.addLeaf(1)
t.addLeaf(2)
t.addLeaf(0)

print(t.calcHeight())
print(maxDepth(t.root))
```
